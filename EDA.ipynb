<h2>Descripción de Variables</h2>
    <ul>
        <li><strong>Gender:</strong> Indica el género del cliente, que puede ser "Male" (Hombre) o "Female" (Mujer). Es una variable categórica que representa el género del cliente.</li>
        <li><strong>Customer Type:</strong> Describe el tipo de cliente, que puede ser "Loyal Customer" (Cliente leal) o "Disloyal Customer" (Cliente desleal). También es una variable categórica que indica la lealtad del cliente.</li>
        <li><strong>Age:</strong> Representa la edad del cliente, que es una variable numérica continua que indica la edad del cliente en años.</li>
        <li><strong>Type of Travel:</strong> Indica el propósito del viaje, que puede ser "Personal Travel" (Viaje personal) o "Business travel" (Viaje de negocios). Es una variable categórica que describe el motivo del viaje.</li>
        <li><strong>Class:</strong> Indica la clase de viaje, que puede ser "Eco" (Económica), "Eco Plus" (Económica Plus) o "Business" (Clase ejecutiva). También es una variable categórica que representa la clase en la que el cliente viajó.</li>
        <li><strong>Flight Distance:</strong> Representa la distancia del vuelo en kilómetros. Es una variable numérica continua que indica la longitud del vuelo.</li>
        <li><strong>Inflight wifi service:</strong> Evalúa el servicio de wifi en el vuelo, donde los valores van de 0 a 5, donde 0 indica un servicio deficiente y 5 indica un servicio excelente.</li>
        <li><strong>Departure/Arrival time convenient:</strong> Evalúa la conveniencia de los horarios de salida/llegada, con valores de 0 a 5, donde 0 es inconveniente y 5 es muy conveniente.</li>
        <li><strong>Ease of Online booking:</strong> Evalúa la facilidad de hacer la reserva en línea, con valores de 0 a 5, donde 0 indica que es difícil y 5 indica que es muy fácil.</li>
        <li><strong>Gate location:</strong> Evalúa la ubicación de la puerta de embarque, con valores de 0 a 5, donde 0 indica una mala ubicación y 5 indica una excelente ubicación.</li>
        <li><strong>Food and drink:</strong> Evalúa la calidad de la comida y bebida a bordo, con valores de 0 a 5, donde 0 indica mala calidad y 5 indica excelente calidad.</li>
        <li><strong>Online boarding:</strong> Evalúa la facilidad del proceso de embarque en línea, con valores de 0 a 5, donde 0 es complicado y 5 es muy fácil.</li>
        <li><strong>Seat comfort:</strong> Evalúa la comodidad de los asientos, con valores de 0 a 5, donde 0 indica asientos incómodos y 5 indica asientos muy cómodos.</li>
        <li><strong>Inflight entertainment:</strong> Evalúa el entretenimiento a bordo, con valores de 0 a 5, donde 0 indica un entretenimiento deficiente y 5 indica un excelente entretenimiento.</li>
        <li><strong>On-board service:</strong> Evalúa el servicio a bordo, con valores de 0 a 5, donde 0 indica un servicio deficiente y 5 indica un excelente servicio.</li>
        <li><strong>Leg room service:</strong> Evalúa el espacio para las piernas, con valores de 0 a 5, donde 0 indica poco espacio y 5 indica mucho espacio.</li>
        <li><strong>Baggage handling:</strong> Evalúa el manejo del equipaje, con valores de 0 a 5, donde 0 indica un manejo deficiente y 5 indica un excelente manejo.</li>
        <li><strong>Checkin service:</strong> Evalúa el servicio de check-in, con valores de 0 a 5, donde 0 indica un servicio deficiente y 5 indica un excelente servicio.</li>
        <li><strong>Inflight service:</strong> Evalúa el servicio a bordo, con valores de 0 a 5, donde 0 indica un servicio deficiente y 5 indica un excelente servicio.</li>
        <li><strong>Cleanliness:</strong> Evalúa la limpieza a bordo, con valores de 0 a 5, donde 0 indica falta de limpieza y 5 indica alta limpieza.</li>
        <li><strong>Departure Delay in Minutes:</strong> Indica la cantidad de minutos de retraso en la salida del vuelo. Es una variable numérica que muestra el retraso en minutos.</li>
        <li><strong>Arrival Delay in Minutes:</strong> Indica la cantidad de minutos de retraso en la llegada del vuelo. Es una variable numérica que muestra el retraso en minutos.</li>
        <li><strong>satisfaction:</strong> Muestra la satisfacción del cliente con su experiencia de vuelo, que puede ser "neutral or dissatisfied" (neutral o insatisfecho) o "satisfied" (satisfecho). Es una variable categórica que refleja la satisfacción del cliente.</li>
    </ul>

```python
sns.countplot(data=df, x='satisfaction')
plt.title('Distribución de la Variable Objetivo')
plt.xlabel('Satisfacción')
plt.ylabel('Frecuencia')
plt.show()

# La proporción de personas insatisfechas es mayor que la de personas satisfechas.
```

```python
sns.countplot(data=df, x='Gender')
plt.title('Distribución de Género')
plt.show()

# En la aerolínea, asisten más mujeres que hombres.
```

```python
sns.countplot(data=df, x='satisfaction', hue='Gender')
# Las mujeres representan el grupo de clientes más insatisfechos.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='satisfaction', y='Age', palette='Set1')
plt.title('Influencia de la Edad en la Satisfacción')
plt.xlabel('Satisfacción')
plt.ylabel('Edad')
plt.xticks([0, 1], ['Insatisfecho', 'Satisfecho'])
plt.show()

#Las personas que tienen edades comprendidas entre los 30 y los 50 años
# muestran niveles más altos de satisfacción.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Class', hue='satisfaction', palette='Set2')
plt.title('Satisfacción por Clase de Viaje')
plt.xlabel('Clase de Viaje')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(title='Satisfacción', loc='upper right', labels=['No Satisfecho', 'Satisfecho'])
plt.show()

#Efectivamente, como se puede observar, los pasajeros de clase Business muestran niveles de 
# satisfacción más altos, seguidos por los pasajeros de clase Eco.
```

```python
satisfaction_mapping = {
    'neutral or dissatisfied': 0,
    'satisfied': 1
}

# # Crear una nueva columna con valores numéricos de satisfacción
df['satisfaction_numeric'] = df['satisfaction'].map(satisfaction_mapping)


numeric_columns = df.select_dtypes(include=['number'])

# Calcula la matriz de correlación
correlation_matrix = numeric_columns.corr()

# Crea el mapa de calor
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, vmin = -1, vmax = 1)
plt.title('Correlación entre Variables Numéricas y Satisfacción')
plt.show()
```

```python
# Estas son algunas de las variables que muestran una correlación ç
# significativa con la satisfacción de los clientes en el conjunto de datos.
threshold = 0.3

# Filtra la matriz de correlación para mostrar solo las correlaciones con satisfacción por encima del umbral
correlation_with_satisfaction = correlation_matrix['satisfaction_numeric']
filtered_correlation = correlation_with_satisfaction[correlation_with_satisfaction.abs() > threshold]

# Crea el mapa de calor solo con las correlaciones filtradas
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_correlation.to_frame(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlación entre Variables Numéricas y Satisfacción (filtrado)')
plt.show()
```
